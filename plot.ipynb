```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f1(x1, x2):
    return 0.5 * (x1**2 + x2**2)

def f2(x1, x2):
    return 0.5 * ((x1 - 1)**2 + (x2 - 1)**2)

# Create a grid of x1 and x2 values
x1 = np.linspace(-2, 3, 400)
x2 = np.linspace(-2, 3, 400)
X1, X2 = np.meshgrid(x1, x2)

# Calculate the function values for each point in the grid
Z1 = f1(X1, X2)
Z2 = f2(X1, X2)

# Create the figure and subplots
fig = plt.figure(figsize=(10, 5))

# Plot the first function
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X1, X2, Z1, cmap='Blues')
ax1.set_title('Function f₁(x₁, x₂) = ½(x₁² + x₂²)')

# Plot the second function
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X1, X2, Z2, cmap='Oranges')
ax2.set_title('Function f₂(x₁, x₂) = ½((x₁ - 1)² + (x₂ - 1)²)')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
f1 = lambda x: 1/2 * (x[0]**2 + x[1]**2)
f2 = lambda x: 1/2 * ((x[0]-1)**2 + (x[1]-1)**2)

# Create a grid of x, y values
x = np.linspace(-2, 3, 400)
y = np.linspace(-2, 3, 400)
X, Y = np.meshgrid(x, y)

# Compute the function values on the grid
Z1 = np.zeros_like(X)
Z2 = np.zeros_like(X)
for i in range(len(x)):
    for j in range(len(y)):
        Z1[i, j] = f1([X[i, j], Y[i, j]])
        Z2[i, j] = f2([X[i, j], Y[i, j]])

# Plot Niveau lines
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.contour(X, Y, Z1, levels=20, colors='blue')
plt.title('$f_1(x)$ Niveau Lines')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')

plt.subplot(1, 2, 2)
plt.contour(X, Y, Z2, levels=20, colors='red')
plt.title('$f_2(x)$ Niveau Lines')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')

plt.tight_layout()
plt.show()
```
